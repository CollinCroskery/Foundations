```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import scipy as sp
from sklearn import linear_model
import math
from scipy.optimize import minimize
from scipy.optimize import Bounds
```

```python
def genSample(iterrations, mean, sd):
 track  = np.empty(((iterrations), 3))
 cumSum = 0   
 for i in range(iterrations):
        track[i, 0] = i+1
        track[i, 1] = np.random.normal(mean, sd, 1)
        cumSum = cumSum + track[i, 1]
        track[i, 2] = cumSum/(i+1)
 return(track)    
```

```python

p1 = go.Figure()
track = genSample(1000, 10, 2)
p1.add_trace(go.Scatter(x = track[:,0], y = track[:,2],
                    mode='lines',
                    name = 'sample 1'                            
                    ))
track = genSample(1000, 10, 2)
p1.add_trace(go.Scatter(x = track[:,0], y = track[:,2],
                    mode='lines',
                    name = 'sample 2'                            
                    ))
track = genSample(1000, 10, 2)
p1.add_trace(go.Scatter(x = track[:,0], y = track[:,2],
                    mode='lines',
                    name = 'sample 3'                            
                    ))
p1.update_layout(autosize = False, width = 500, height = 400, 
                 plot_bgcolor = '#FFFFFF')

p1.show()
```

```python
data = np.random.normal(10, 2, 1000)
```

```python
def llNorm(parameters):
 mu = parameters[0]
 sigma = parameters[1]
 NLL = -len(data)/2 * np.log(2 * np.pi) - (len(data)/2) * np.log(sigma**2)
 tmp = 0
 for val in data:
  tmp = tmp + ((val-mu)**2)
 NLL = NLL + -(1/(2*(sigma**2)))*tmp
 return(-NLL) 

def proposalfunction(param, t):
 return(float(np.random.uniform((param - t),(param + t),1))) # be more careful with Python more stongly typed

def MCMC(startval, iterrations, t):
#    chain  = np.zeros(((iterrations+1), 2))
    chain  = np.empty(((iterrations+1), 2))
    chain[0,1] = startval
    sigma = 2 
    for i in range(iterrations):
     chain[i, 0] = i
     proposal = proposalfunction(chain[i,1], t)
     if (llNorm(np.array([chain[i,1],sigma])) < llNorm(np.array([proposal,sigma]))):
        chain[i+1,1] = chain[i,1]
     else:
        chain[i+1,1] = proposal
    return(chain)
    
```

```python
iterations = 100
p1 = go.Figure()
chain = MCMC(6, iterations, 1)
p1.add_trace(go.Scatter(x = chain[:,0], y = chain[:,1],
                    mode='lines',
                    name = 'sample 1'                            
                    ))
chain = MCMC(11, iterations, 1)

p1.add_trace(go.Scatter(x = chain[:,0], y = chain[:,1],
                    mode='lines',
                    name = 'sample 2'                            
                    ))
chain = MCMC(20, iterations, 1)
p1.add_trace(go.Scatter(x = chain[:,0], y = chain[:,1],
                    mode='lines',
                    name = 'sample 3'                            
                    ))
p1.update_layout(autosize = False, width = 500, height = 400, 
                 plot_bgcolor = '#FFFFFF')

p1.show()
```
