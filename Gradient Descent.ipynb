```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import scipy as sp
from sklearn import linear_model
import math
```

```python
Intercept = np.array([3.5])
b_1 = np.array([1.4])
X = np.array([3])
y = np.array([Intercept + (b_1 * X)])

theta = np.array([1.4])

def cost2(X, y, theta):
  return(np.sum((((X.dot(theta) + Intercept) - y)**2))/(2*y.size))
print (X)

# np.sum((((X.dot(theta) + Intercept) - y)**2))/(2*y.size)
```

```python
alpha = 0.02 
# alpha controls the learning rate (step size) 
# try 0.2 (and note the spaces and how it jumps side to side 
# and also 0.005 wich is more typical of relative step 

num_iters = 100
theta = 0
#chartPts = NULL need to build array differently

epsilon = .001 
# epsion controls the threshold - try 2 (note how it's likely to totally miss
# or .01
```

```python

chartPts = np.empty((0,2))
tracerMat = np.empty((0,6))

tracerMat2 = np.empty((0,5))

# i = 1

for i in range(num_iters):
    
 #  error <- ((X %*% theta)+Intercept - y)
 error =  (X.dot(theta) + Intercept) - y
 # we keep the sign here bc it detemines the direction of the descent
 delta = (np.transpose(X).dot(error)) /y.size
 theta = theta - (alpha * delta)
 cost = cost2(X, y, theta)
 chartPts = np.append(chartPts, np.array([[float(theta), cost]]), axis=0)
 if ((math.sqrt(np.sum(delta**2)))<=epsilon): 
  break 
 tracerMat = np.append(tracerMat, np.array([[int(i), float(error), float(delta), float(theta), cost, math.sqrt(np.sum(delta**2))]]), axis=0)  
```
